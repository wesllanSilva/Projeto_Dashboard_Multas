<h1>Importar os Pacotes/Bibliotecas/Libs</h1>

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

# Acessar compenentes do SISTEMA OPERACIONAL
import os
from pathlib import Path
```

<h1>Localizar diretorio diretorio no windows</h1>

```python
Local_Diretorio = 'C:/WORKSPACE/Projeto_Dashboard_Multas/infrações_2022/ajustados_2022'
Diretorio_Path = Path(Local_Diretorio)

type(Diretorio_Path)
```

<h1>Listar e interagir com os Arquivos</h1>

```python
# Listar os arquivos
# variavel para interagir com os arquivos
Arquivos = Diretorio_Path.iterdir()

# Função para calcular o tamanho do arquivo
def Bytes_Mega(bytes):
    Megabytes = bytes / ( 1024 * 1024)
    return Megabytes


# Total MBs
Total_MB = 0

# Loop
for Loop in Arquivos:
    print(Loop)
    # Buscando o tamanho do arquivo
    Tamanho = os.path.getsize( Loop )
    
    # Convervento em Megas
    Conversao = Bytes_Mega( Tamanho )
    
    print(Conversao)
```

<h1>Melhorando a visualização</h1>  <h3>Minerando informação de texto</h3>

```python
Arquivos = Diretorio_Path.iterdir()

def Bytes_Mega(bytes):
    Megabytes = bytes / ( 1024 * 1024)
    return Megabytes

Total_MB = 0

# Loop
for Loop in Arquivos:
    
    Tamanho = os.path.getsize( Loop )
    
    Conversao = Bytes_Mega( Tamanho )
    
    # Extraindo o nome do arquivo e minerando informação de texto.
    Nome_Arquivo = str( Loop ).split('\\')[-1]
    
    
    print(f' {Nome_Arquivo} : { round(Conversao, 2) } MB')
    print('-' * 30 )
    
    
    # Tamanho total dos arquivos
    Total_MB = Total_MB + Tamanho
    
print( f'Temos ao todo { round( Bytes_Mega(Total_MB), 2) } MBs de dados' )
```

<h1> Consolidando os dados</h1>

```python
# Lista os arquivos
Arquivos = Diretorio_Path.iterdir()

# Loop para consolidar os arquivos
for Contador, Arq in enumerate(Arquivos):
    
    print(Contador, Arq )
    
    # Se for a 1º interação, cria a base consolidada
    if Contador == 0:
        # Ler o arquivo do loop
        Base_Dados = pd.read_csv( Arq, error_bad_lines=False, sep=';', encoding='latin-1' )
        # Salvar na variavel 
        Base_Consolidada = Base_Dados
        
    else:
        # Ler o arquivo do loop
        Base_Dados = pd.read_csv( Arq, error_bad_lines=False, sep=';', encoding='latin-1' )
        # Empilhar os dados
        Base_Consolidada = pd.concat( [Base_Consolidada, Base_Dados] )
    
# Dimensão
Base_Consolidada.shape
```

<h1> Exportação para .CSV </h1>

```python
Base_Consolidada.to_csv('Base_Consolidada.csv', index=False)
```

<h2>Para resolver o problema de conversão para parquet, é preciso converter as colunas para formato texto ('str')</h2>
